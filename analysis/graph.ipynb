```python
import numpy as np 
import matplotlib.pyplot as plt
import os.path
```

```python
font = {
    'weight' : 'normal',
    'size' : 12
}
```

### Prunning with transfer learning 

```python
Sparsity = ['dense', '50%', '70%', '80%', '90%']
Res34_cifar10 = [96.130, 96.580, 96.370, 96.200, 95.780]
Res34_food101 = [82.178, 82.222, 82.606, 81.406, 81.667]
Vgg16_cifar10 = [93.080, 93.300, 93.810, 94.260, 93.330]
Vgg16_food101 = [78.547, 80.602, 80.432, 80.135, 77.085]

plt.tick_params(labelsize=10)
plt.plot(Sparsity, Res34_cifar10, color='red', marker='o', linestyle='-', label='Res34->Cifar10')
plt.plot(Sparsity, Vgg16_cifar10, color='blue', marker='o', linestyle='-',label='Vgg16->Cifar10')
plt.plot(Sparsity, Res34_food101, color='red', marker='*', linestyle=':', label='Res34->Food101')
plt.plot(Sparsity, Vgg16_food101, color='blue', marker='*', linestyle=':',label='Vgg16->Food101')
plt.legend(loc='center right')
plt.title('Transfer learning with pruned models', fontdict=font)
plt.xlabel = ('Sparsity%')
plt.ylabel =('TOP1 Accuracy%')
filename = os.path.join(dir,"prunning_transferlearning.png")
plt.savefig(filename, dpi=300)
plt.show()
```

### Resnet34(Imagenet Pre-trained) agp pruning to correspond sparsity then conduct transfer learning on Cifar10_resize dataset.(224x224)  vs transfering learning with dense model then prune to correspond sparsity

```python
Sparsity = ['dense', '50%', '70%', '80%', '90%']
Prune1Accuracy = [96.130, 96.580, 96.370, 96.200, 95.780]
Transfer1Accuracy = [96.130, 96.450, 96.610, 96.570, 96.300]
plt.plot(Sparsity, Prune1Accuracy, color='red', marker='o', linestyle='solid', label='P-T')
plt.plot(Sparsity, Transfer1Accuracy, color='blue', marker='o', linestyle='solid',label='T-P')
plt.legend(loc='lower left')
plt.title('Resnet34 pruning and transfer learning on Cifar10', fontdict=font)
plt.xlabel = ('Sparsity%')
plt.ylabel =('TOP1 Accuracy%')
filename = os.path.join(dir,"Resnet34_Cifar10.png")
#plt.tight_layout()
#plt.subplots_adjust(bottom=0.1)
plt.savefig(filename, dpi=300)
plt.show() 
```

#### Resnet34 and Vgg16 pruning before/after transfer learning on cifar10

```python
Sparsity = ['dense', '50%', '70%', '80%', '90%']
Res34_P2T = [96.130, 96.580, 96.370, 96.200, 95.780]
Res34_T2P = [96.130, 96.450, 96.610, 96.570, 96.300]
Vgg16_P2T = [93.080, 93.300, 93.810, 94.260, 93.330]
Vgg16_T2P = [93.080, 94.560, 95.070, 94.680, 94.600]

plt.plot(Sparsity, Res34_P2T, color='red', marker='o', linestyle='-', label='Res34_PB')
plt.plot(Sparsity, Res34_T2P, color='blue', marker='o', linestyle='-',label='Res34_PA')

plt.plot(Sparsity, Vgg16_P2T, color='red', marker='*', linestyle=':', label='Vgg16_PB')
plt.plot(Sparsity, Vgg16_T2P, color='blue', marker='*', linestyle=':',label='Vgg16_PA')

plt.title("Resnet34 and Vgg16 pruning before/after transfer learning on Cifar10")
plt.ylabel =("TOP1 Accuracy")
plt.xlabel = ("Sparsity")


plt.legend(loc='best', fontsize=8)
filename = os.path.join(dir,"Res34_Vgg16_cifar10.png")
plt.savefig(filename, dpi=300)
plt.show()
```

### Resnet34(Imagenet Pre-trained) agp pruning to correspond sparsity then conduct transfer learning on Cifar10_resize 224x224)  then further prune the model to correspond sparsity 

```python
spa0_trans_prune_spa = ['dense', '50%', '70%', '80%', '90%']
spa0_trans_prune_acc = [96.130, 96.580, 96.370, 96.200, 95.780]
spa50_trans_prune_spa = ['50%', '70%', '80%', '90%']
spa50_trans_prune_acc = [96.580, 96.650, 96.740, 96.500]
spa70_trans_prune_spa = ['70%', '80%', '90%']
spa70_trans_prune_acc = [96.370, 96.610, 96.350]
spa80_trans_prune_spa = ['80%', '90%']
spa90_trans_prune_acc = [96.200, 96.290]

Res34_T2P = [96.130, 96.450, 96.610, 96.570, 96.300]

plt.ylabel =["TOP1 Accuracy"]
plt.xlabel = ["Sparsity"]
plt.plot(spa0_trans_prune_spa, spa0_trans_prune_acc, color='black', marker='o', linestyle=':', label='PB')
plt.plot(spa0_trans_prune_spa, Res34_T2P, color='red', marker='o', linestyle=':', label='PA')

plt.plot(spa50_trans_prune_spa, spa50_trans_prune_acc, color='orange', marker='o', linestyle='-', label='PBA 50%')
plt.plot(spa70_trans_prune_spa, spa70_trans_prune_acc, color='green', marker='o', linestyle='-',label='PBA 70%')
plt.plot(spa80_trans_prune_spa, spa90_trans_prune_acc, color='blue', marker='o', linestyle='-',label='PBA 80%')
plt.title("Resnet34 pruning and transfer learning on Cifar10")
plt.legend(loc='lower left')
filename = os.path.join(dir,"Resnet34_PTP_Cifar10.png")
plt.savefig(filename, dpi=300)
plt.show()
```
